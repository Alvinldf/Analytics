# Trabajando con fechas

```python
fallecidos_covid['Fe_FALLECIMIENTO'] = pd.to_datetime(fallecidos_covid.FECHA_FALLECIMIENTO, format = '%Y%m%d')
```

```python
fallecidos_covid
```

```python
fallecidos_covid['year'] = pd.DatetimeIndex(fallecidos_covid.Fe_FALLECIMIENTO).year
fallecidos_covid['month'] = pd.DatetimeIndex(fallecidos_covid.Fe_FALLECIMIENTO).month
fallecidos_covid['day'] = pd.DatetimeIndex(fallecidos_covid.Fe_FALLECIMIENTO).day
fallecidos_covid['weekday'] = pd.DatetimeIndex(fallecidos_covid.Fe_FALLECIMIENTO).weekday
```

```python
fallecidos_covid
```

```python
fallecidos_covid[fallecidos_covid.month==12]
```

```python
muertes_por_mes = fallecidos_covid.groupby('month')['UUID'].count()
muertes_por_mes.sort_values(ascending=False)
```

```python
fallecidos_covid.sort_values('Fe_FALLECIMIENTO', ascending=False)
```

```python
muertes_por_departamento= fallecidos_covid.groupby('DEPARTAMENTO')['UUID'].count()
muertes_por_departamento.sort_values(ascending=False)
```

```python
muertes_por_dia = fallecidos_covid.groupby('Fe_FALLECIMIENTO')['UUID'].count()
muertes_por_dia
```

```python
muertes_por_dia.describe()
```

```python
muertes_por_dia.plot();
```

```python
sns.set_style("darkgrid")
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (14, 6)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
plt.plot(muertes_por_dia, 'r')
plt.title('Muertes diarias por Covid_19 - Perú ')
plt.xlabel('Fecha')
plt.ylabel('Número de muertes');
```
